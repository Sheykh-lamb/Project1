### 2.Задачи на логику

#### Задача 1

1.Рекламная кампания стартовала вчера с дневным бюджетом 40 $. Половина
бюджета была израсходована к полудню, а 80% оставшегося бюджета было
потрачено между полуднем и временем закрытия. Сколько долларов не было
потрачено?

Найдем бюджет к полудню:
$$ x = \frac{40}{2} = 20 $$

Далее используем пропорцию

20 = 100%

x = 80%

$$ 20 * 80 = 100 * x $$

$$ x = \frac{20 * 80}{100} = 16 $$

$$ 20 - 16 = 4$$

**Ответ: Не было потрачено 4 доллара.**

#### Задача 2

5 идентичных рекламных кампаний работали 24 дня по 6 часов в день,
потрачено было 120 долларов. Сколько дней они работали бы на 216 долларов,
если бы 9 одинаковых кампаний работали бы по 8 часов в день?

Узнаем сколько одна компания заработала за 24 дня:

$$ \frac{120}{5} = 24$$

Узнаем сколько зарабатывает одна компания в час:

$$ \frac{24}{24 * 6} = 0.17$$

На основе полученных данных мы можем узнать, сколько дней проработало бы 9 компаний за 216 долларов:

$$ \frac{216}{9*8*0.17} = 17,65$$

Вышло 17.65, однако день не может быть разбит на части, поэтому есть два выхода из ситуации:
1) Если компании не готовы пойти на снижении оплаты за последний неполный день, то округляем в меньшую сторону - 17 дней.
2) В ином случае округляем в большую сторону - 18 дней.

**Ответ: В зависимости от ситуации ответ 17 или 18 дней**

#### Задача 3

Дизайнеры создали 200 рекламных баннеров для двух рекламных кампаний. 80
из них не использовались ни в кампании №1, ни в кампании №2, 60
использовались только в кампании №1. И для каждого рекламного баннера,
который использовался в обеих кампаниях, приходится 3 баннера, которые
использовались только в кампании №2. Сколько баннеров было использовано в
обеих рекламных кампаниях?

Сперва узнаем сколько всего использовали баннеров:

$$ 200 - 80 = 120$$

Далее узнаем сколько было всего баннеров в кампании №2 и в обеих кампаниях:

$$ 120 - 60 = 60 $$

Количество использованных баннеров в обеих компаниях обозначим как `x`, тогда в кампании №2 было `3x` баннеров.

$$ x + 3x = 60 $$

$$ x = \frac{60}{4} = 15$$

**Ответ: В обеих кампаниях было использовано 15 баннеров.**

#### Задача 5

Средний итоговый балл стажера по 4 модулям составляет 78 баллов. Сколько
баллов должен получить стажер за 5-й модуль, чтобы средний балл по всему
заданию составил 80?

Обозначим:

`x - количество баллов за первые 4 модуля`

`y - необходимое количество баллов за 5 модуль`

Узнаем сколько баллов получил студент за первые 4 модуля:

$$\frac{x}{4} = 78$$

$$x = 312$$

За 4 модуля стажер получил 312 баллов, на основе этого мы можем узнать, сколько ему ещё баллов нужно для среднего балла 80:


$$\frac{312 + y}{5} = 80$$

$$y = 88$$

**Ответ**: За 5 модуль необходимо получить 88 баллов.

#### Задача 6 

Заказанные в приложении товары доставляются на автомобиле. Автомобиль
проезжает 260 км со средней скоростью 80 км / ч. На обратном пути машина
движется со средней скоростью 100 км / ч. Насколько быстрее был обратный
путь? Ответ указать в минутах

Сперва найдем время, за которое автомобиль в первый раз проехал путь:

$$ \frac{260}{80} = 3.25$$

Со средней скоростью 80 км / ч автомобиль проехал путь за 3.25 часа.

Найдем время за которое он проедет обратный путь:

$$ \frac{260}{100} = 2.6$$

Со средней скоростью 100 км / ч автомобиль проехал путь за 2.6 часа.

На основе полученных данных мы можем узнать, насколько быстрее был проделан путь обратно:

$$ 3.25 - 2.6 = 0.65$$

Так как время нужно в минутах умножим на 60 

$$ 60 * 0.65 = 39$$

**Ответ:** Обратный путь был проделан на 39 минут быстрее.
